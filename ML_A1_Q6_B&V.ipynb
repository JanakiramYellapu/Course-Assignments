```
import numpy as np
import math
import matplotlib.pyplot as plt
```

```
def generate_data(x):
  y = []
  for i in x:
    y.append( math.cos(2*math.pi*i) + math.tanh(2*math.pi*i))
  return y
```

```
#  Generating data domain(0,1)

data = np.random.uniform(0, 1, 100)
train_x = data[0:80]
test_x = data[80:100]
train_size = len(train_x)


# Function
test_y = []
train_y = []
test_y = generate_data(test_x)
# for i in range(train_size):
#   train_y.append( math.cos(2*math.pi*train_x[i]) + math.tan(2*math.pi*train_x[i]))
train_y = generate_data(train_x)
#plotting
plt.plot(train_x, train_y,'bo') 
plt.show()

# Plottin after adding noise (0, 0.2)
mean = 0
std = np.sqrt(0.2)
y_noise = train_y + np.random.normal(mean, std, train_size)
y_noise_test = test_y + np.random.normal(mean, std, 20)
plt.plot(train_x, y_noise,'bo') 
plt.show()
```

```
# Polynomial Regression for 10 points
degrees = [1, 3, 6, 9]
# degrees = [9]
n = [10, 20, 40, 80]

# Test data in test_x and y_noise_test


for d in degrees:
  rms_test = []
  rms_train = []
  for i in range(len(n)):
    x_10 = train_x[:n[i]]
    y_10 = y_noise[:n[i]]

    model = np.poly1d(np.polyfit(x_10, y_10, d))
    line = np.linspace(x_10.min(), x_10.max())
    # plt.scatter(x_10, y_10)
    # plt.plot(line, model(line))
    # plt.title("Polynomial Regression degree : %i " %(d))
    # plt.xlabel("Input (x)")
    # plt.ylabel("Output (y)")
    # plt.show()
    # rms_train.append(np.sqrt(np.sum((model(x_10) - y_10)**2)/n[i]))
    rms_test.append(np.sqrt(np.sum((model(test_x) - y_noise_test)**2)/n[i]))

  plt.plot(n,rms_test)
  plt.xlabel("N(no of samples)")
  plt.ylabel("Testing error")
  plt.title("N vs Testing error for degree %d " %(d))
  # plt.plot(n,rms_train)
  plt.show()



```
